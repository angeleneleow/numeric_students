# Problem RK4

```python
import context
import numpy as np
from matplotlib import pyplot as plt
from numlabs.lab4.lab4_functions import initinter41,eulerinter41,midpointinter41,\
                                        rk4ODEinter41                                    
initialVals={'yinitial': 1,'t_beg':0.,'t_end':1.,'dt':0.05,'c1':-1.,'c2':1.,'c3':1.}
coeff = initinter41(initialVals)
timeVec=np.arange(coeff.t_beg,coeff.t_end,coeff.dt)
nsteps=len(timeVec)
ye=[]
ym=[]
yrk=[]
y=coeff.yinitial
ye.append(coeff.yinitial)
ym.append(coeff.yinitial)
yrk.append(coeff.yinitial)
for i in np.arange(1,nsteps):
    ynew=eulerinter41(coeff,y,timeVec[i-1])
    ye.append(ynew)
    ynew=midpointinter41(coeff,y,timeVec[i-1])
    ym.append(ynew)
    ynew=rk4ODEinter41(coeff,y,timeVec[i-1])
    yrk.append(ynew)
    y=ynew
analytic=timeVec + np.exp(-timeVec)
theFig=plt.figure(0)
theFig.clf()
theAx=theFig.add_subplot(111)
l1=theAx.plot(timeVec,analytic,'b-',label='analytic')
theAx.set_xlabel('time (seconds)')
l2=theAx.plot(timeVec,ye,'r-',label='euler')
l3=theAx.plot(timeVec,ym,'g-',label='midpoint')
l4=theAx.plot(timeVec,yrk,'m-',label='rk4')
theAx.legend(loc='best')
theAx.set_title('interactive 4.2');
```

## Accuracy

At the scale presented in graph all methods appear to have similar accuracy. However, from below, rk4 is more accurate for $dt=0.05$

```python
n = np.arange(0,4)
dt_n = 0.05/(2**n)
ye_err=[]
ym_err=[]
yrk_err=[]
for dt in dt_n:
    initialVals={'yinitial': 1,'t_beg':0.,'t_end':1.,'dt':dt,'c1':-1.,'c2':1.,'c3':1.}
    coeff = initinter41(initialVals)
    timeVec=np.arange(coeff.t_beg,coeff.t_end,coeff.dt)
    nsteps=len(timeVec)
    ye=[]
    ym=[]
    yrk=[]
    y=coeff.yinitial
    ye.append(coeff.yinitial)
    ym.append(coeff.yinitial)
    yrk.append(coeff.yinitial)
    for i in np.arange(1,nsteps):
        ynew=eulerinter41(coeff,y,timeVec[i-1])
        ye.append(ynew)
        ynew=midpointinter41(coeff,y,timeVec[i-1])
        ym.append(ynew)
        ynew=rk4ODEinter41(coeff,y,timeVec[i-1])
        yrk.append(ynew)
        y=ynew
    analytic=timeVec + np.exp(-timeVec)
    ye_err.append(np.sqrt(np.mean((ye-analytic)**2)))
    ym_err.append(np.sqrt(np.mean((ym-analytic)**2)))
    yrk_err.append(np.sqrt(np.mean((yrk-analytic)**2)))
theFig=plt.figure(0)
theFig.clf()
theAx=theFig.add_subplot(111)
theAx.set_xlabel('dt')
theAx.set_ylabel('RMSE')
theAx.plot(dt_n,ye_err,'r-',label='euler')
theAx.plot(dt_n,ym_err,'g-',label='midpoint')
theAx.plot(dt_n,yrk_err,'m-',label='rk4')
theAx.legend(loc='best')
theAx.set_title('Error');
theAx.invert_xaxis(); 
```

# Problem coding A

```python
import json
from numlabs.lab4.example.do_example import get_init
#
# specify the derivs function
#
def derivs(coeff, y):
  f=np.empty_like(y) #create a 2 element vector to hold the derivitive
  f[0]=y[1]
  f[1]= -1.*coeff.c1*y[1] - coeff.c2*y[0]
  return f

def Heuns(coeff,y,derivs):
    k1 = coeff.dt * derivs(coeff,y)
    k2 = coeff.dt * derivs(coeff,y + (2/3 * k1))
    f = y + (1/4)*k1 + (3/4)*k2
    return f
#
# first make sure we have an input file in this directory
#

coeff=get_init()

#
# integrate and save the result in savedata
#
time=np.arange(coeff.t_beg,coeff.t_end,coeff.dt)
y=coeff.yinitial
nsteps=len(time) 
savedata=np.empty([nsteps],np.float64)
for i in range(nsteps):
    y=Heuns(coeff,y,derivs)
    savedata[i]=y[0]

theFig,theAx=plt.subplots(1,1,figsize=(8,8))
theAx.plot(time,savedata,'o-')
theAx.set_title(coeff.plot_title)
theAx.set_xlabel('time (seconds)')
theAx.set_ylabel('y0');

```

# Problem coding B

```python
from numlabs.lab4.lab4_functions import initinter41, derivsinter41, midpointinter41


def heunsinter41(coeff,y,theTime):
    k1 = coeff.dt * derivsinter41(coeff,y,theTime)
    k2 = coeff.dt * derivsinter41(coeff,y + (2/3 * k1),theTime + (2/3 * coeff.dt))
    f = y + (1/4)*k1 + (3/4)*k2
    return f

initialDict={'yinitial': 1,'t_beg':0.,'t_end':1.,
                    'dt':0.1,'c1':-1.,'c2':1.,'c3':1.}    
coeff = initinter41(initialDict)

time=np.arange(coeff.t_beg,coeff.t_end,coeff.dt)
nsteps=len(time) 
y_h=np.empty(nsteps)
y_m=np.empty(nsteps)
y_h[0]=coeff.yinitial
y_m[0]=coeff.yinitial
savedata_h=np.empty([nsteps],np.float64)
savedata_m=np.empty([nsteps],np.float64)
for i in range(1,nsteps):
    y_h[i]=heunsinter41(coeff,y_h[i-1],time[i])
    y_m[i]=midpointinter41(coeff,y_m[i-1],time[i])
    
theFig,theAx=plt.subplots(1,1,figsize=(8,8))
theAx.plot(time,y_h,'o-',label="Heun's")
theAx.plot(time,y_m,'o-',label="Midpoint")
theAx.set_title('y(0) = 1')
theAx.set_xlabel('time (seconds)')
theAx.set_ylabel('y');
theAx.legend()
```

Both methods are the same in the above examples. This can be shown as writing both methods as an explicit two-stage RK method of the form 

\begin{aligned}
&k_{1}=h f\left(y_{n}, t_{n}\right)\\
&k_{2}=h f\left(y_{n}+a k_{1}, t_{n}+a h\right)\\
&y_{n+1}=y_{n}+c_{1} k_{1}+c_{2} k_{2}
\end{aligned}

Where the mid point method is 

\begin{aligned}
&k_{1}=h f\left(y_{n}, t_{n}\right)\\
&k_{2}=h f\left(y_{n}+\frac{1}{2} k_{1}, t_{n}+\frac{1}{2} h\right)\\
&y_{n+1}=y_{n}+k_{2}
\end{aligned}

and Heun's method is 

\begin{aligned}
&k_{1}=h f\left(y_{n}, t_{n}\right)\\
&k_{2}=h f\left(y_{n}+\frac{2}{3} k_{1}, t_{n}+\frac{2}{3} h\right)\\
&y_{n+1}=y_{n}+\frac{1}{4} k_{1}+\frac{3}{4} k_{2}
\end{aligned}

Thus for $f(y, t)=t-y+1$, if we can show that $k_{2} = \frac{1}{4} k_{1}^*+\frac{3}{4} k_{2}^*$ (where $^*$ is denotes the coefficients in Heun's method), both methods are equivalent.


Expanding the right side we get,


$$
\frac{1}{4} k_{1}^*+\frac{3}{4} k_{2}^* = \frac{1}{4}\left( t-y+1 \right) + \frac{3}{4}\left( t+\frac{2}{3}h-\left(y+\frac{2}{3}\left( t-y+1 \right)\right)+1\right) = \frac{1}{2}t+\frac{1}{2}h-\frac{1}{2}y+\frac{1}{2}
$$

Expanding out the left side we get,

$$
k_2 = t+\frac{1}{2}h-\left(y+\frac{1}{2}\left(t-y+1\right)\right)+1 = \frac{1}{2}t+\frac{1}{2}h-\frac{1}{2}y+\frac{1}{2} =\frac{1}{4} k_{1}^*+\frac{3}{4} k_{2}^*
$$

Therefore, for $f(y, t)=t-y+1$, The midpoint and Heun's method are the same.
