### problem 2 (tableau for Heun-Method)

Heun's method is a second-order method with two stages.


$$
  \begin{array}{l}
    k_1 =  h f(y_n,t_n)\\
    k_2 = h f(y_n+\frac{2}{3}k_1, t_n+\frac{2}{3}h)\\
    y_{n+1} = y_n + \frac{1}{4}k_1 + \frac{3}{4}k_2
  \end{array}
$$

The coefficients can be expressed in a tabular form.

$$
\begin{array}{|c c|cc|c|} \hline
i & a_i &{b_{ij}} && c_i \\ \hline
1 & a_1 & b_{11} & b_{12}  & \frac{1}{4}\\
2 & a_2 & b_{21} & b_{22} & \frac{3}{4}\\\hline 
{j=} & & 1 & 2  & \\ \hline
\end{array}
$$

$$
\begin{array}{|c c|cc|c|} \hline
i & a_i &{b_{ij}} && c_i \\ \hline
1 & 0 & 0 & 0  & \frac{1}{4}\\
2 & \frac{2}{3} & \frac{2}{3} & 0 & \frac{3}{4}\\\hline 
{j=} & & 1 & 2  & \\ \hline
\end{array}
$$


### problem 2 (tableau for 4th order runge-kutta)
A 4-stage Runge-Kutta method formula is:


$$
  \begin{array}{l}
    k_1 =  h f(y_n,t_n)\\
    k_2 = h f(y_n+\frac{k_1}{2}, t_n+\frac{h}{2})\\
    k_3 = h f(y_n+\frac{k_2}{2}, t_n+\frac{h}{2})\\
    k_4 = h f(y_n+k_3, t_n+h)\\
    y_{n+1} = y_n + \frac{k_1}{6}+ \frac{k_2}{3}+ \frac{k_3}{3} + \frac{k_4}{6}
  \end{array}
$$

The coefficients can be expressed in a tabular form known as the
Runge-Kutta tableau. 

$$
\begin{array}{|c|c|cccc|c|} \hline
i & a_i &{b_{ij}} &  &  && c_i \\ \hline
1 & a_1 & b_{11} & b_{12} & b_{13} & b_{14} & c_1\\
2 & a_2 & b_{21} & b_{22} & b_{23} & b_{24} & c_2\\ 
3 & a_3 & b_{31} & b_{32} & b_{33} & b_{34} & c_3\\ 
4 & a_4 & b_{41} & b_{42} & b_{43} & b_{44} & c_4\\\hline 
{j=} & & 1 & 2 & 3 & 4 & \\ \hline
\end{array}
$$

$$
\begin{array}{|c|c|cccc|c|} \hline
i & a_i &{b_{ij}} &  &  && c_i \\ \hline
1 & 0 & 0 & 0 & 0 & 0 & \frac{1}{6}\\
2 & \frac{1}{2} & \frac{1}{2} & 0 & 0 & 0 & \frac{1}{3}\\ 
3 & \frac{1}{2} & 0 & \frac{1}{2} & 0 & 0 & \frac{1}{3}\\ 
4 & 1 & 0 & 0 & 1 & 0 & \frac{1}{6}\\\hline 
{j=} & & 1 & 2 & 3 & 4 & \\ \hline
\end{array}
$$

### problem 3:RK 4 

$$
\frac{dy}{dt} = -y +t +1,  \;\;\;\; y(0) =1
$$ 

(1) Based on the numerical solutions of (eq:test), which of the three methods appears more accurate?

(2) Again determine how the error changes relative to the change in stepsize, as the stepsize is halved.

```python
# problem 3 (1)
initialVals={'yinitial': 1,'t_beg':0.,'t_end':1.,'dt':0.05,'c1':-1.,'c2':1.,'c3':1.}
coeff = initinter41(initialVals)
timeVec=np.arange(coeff.t_beg,coeff.t_end,coeff.dt)
nsteps=len(timeVec)
ye=[]
ym=[]
yrk=[]
y=coeff.yinitial
ye.append(coeff.yinitial)
ym.append(coeff.yinitial)
yrk.append(coeff.yinitial)
for i in np.arange(1,nsteps):
    ynew=eulerinter41(coeff,y,timeVec[i-1])
    ye.append(ynew)
    ynew=midpointinter41(coeff,y,timeVec[i-1])
    ym.append(ynew)
    ynew=rk4ODEinter41(coeff,y,timeVec[i-1])
    yrk.append(ynew)
    y=ynew
    
analytic=timeVec + np.exp(-timeVec)
theAx = plt.figure(figsize = (10,10))
theAx=plt.subplot(211)
l1=theAx.plot(timeVec,analytic,'b-',label='analytic')
theAx.set_xlabel('time (seconds)')
l2=theAx.plot(timeVec,ye,'r-',label='euler')
l3=theAx.plot(timeVec,ym,'g-',label='midpoint')
l4=theAx.plot(timeVec,yrk,'m-',label='rk4')
theAx.legend(loc='best')
theAx.set_title('numerical methods');

error_e = np.absolute(ye-analytic)
error_m = np.absolute(ym-analytic)
error_rk = np.absolute(yrk-analytic)
newAx=plt.subplot(212)
l2=newAx.plot(timeVec,error_e,'r-',label='euler')
l3=newAx.plot(timeVec,error_m,'g-',label='midpoint')
l4=newAx.plot(timeVec,error_rk,'m-',label='rk4')
newAx.set_xlabel('time (seconds)')
newAx.legend()
newAx.set_title("Error for dt = 0.05")

print('Error_rk' +str(error_rk))
```

Based on the first subplot, we see that all methods have the same accuracy. However, the the second error subplot shows that rk4 is more accurate as it is closest to zero. We can analyze further with a halved timestep.

```python
# problem 3 (2)
initialVals={'yinitial': 1,'t_beg':0.,'t_end':1.,'dt':0.0025,'c1':-1.,'c2':1.,'c3':1.}
coeff = initinter41(initialVals)
timeVec=np.arange(coeff.t_beg,coeff.t_end,coeff.dt)
nsteps=len(timeVec)
ye=[]
ym=[]
yrk=[]
y=coeff.yinitial
ye.append(coeff.yinitial)
ym.append(coeff.yinitial)
yrk.append(coeff.yinitial)
for i in np.arange(1,nsteps):
    ynew=eulerinter41(coeff,y,timeVec[i-1])
    ye.append(ynew)
    ynew=midpointinter41(coeff,y,timeVec[i-1])
    ym.append(ynew)
    ynew=rk4ODEinter41(coeff,y,timeVec[i-1])
    yrk.append(ynew)
    y=ynew
    
theAx = plt.figure(figsize = (10,10))
error_e = np.absolute(ye-analytic)
error_m = np.absolute(ym-analytic)
error_rk = np.absolute(yrk-analytic)
newAx=plt.subplot(212)
l2=newAx.plot(timeVec,error_e,'r-',label='euler')
l3=newAx.plot(timeVec,error_m,'g-',label='midpoint')
l4=newAx.plot(timeVec,error_rk,'m-',label='rk4')
newAx.set_xlabel('time (seconds)')
newAx.legend()
newAx.set_title("Error for dt = 0.025" )
```

>Based on dt= 0.025, we see that rk4 and midpoint is most accurate. The error generated when dt = 0.025 is smaller than when dt = 0.05. This tells us that the smaller the time step, the more accurate the measurement given using rk4 and midpoint methods.

### Problem 4 (Embedded Runge-Kutta Methods)

```python
from numlabs.lab4.lab4_functions import initinter41,rk4ODEinter41,rkckODEinter41
initialVals={'yinitial': 1,'t_beg':0.,'t_end':1.,'dt':0.2,'c1':-1.,'c2':1.,'c3':1.}
coeff = initinter41(initialVals)

timeVec=np.arange(coeff.t_beg,coeff.t_end,coeff.dt)
nsteps=len(timeVec)
ye=[]
ym=[]
yrk=[]
yrkck=[]
y1=coeff.yinitial
y2=coeff.yinitial
yrk.append(coeff.yinitial)
yrkck.append(coeff.yinitial)
for i in np.arange(1,nsteps):
    ynew=rk4ODEinter41(coeff,y1,timeVec[i-1])
    yrk.append(ynew)
    y1=ynew 
    ynew=rkckODEinter41(coeff,y2,timeVec[i-1])
    yrkck.append(ynew)
    y2=ynew 
analytic=timeVec + np.exp(-timeVec)
theAx = plt.figure(figsize = (10,10))
theAx=plt.subplot(211)
l1=theAx.plot(timeVec,analytic,'b-',label='analytic')
theAx.set_xlabel('time (seconds)')
l2=theAx.plot(timeVec,yrkck,'g-',label='rkck')
l3=theAx.plot(timeVec,yrk,'m-',label='rk')
theAx.legend(loc='best')
theAx.set_title('rk4 vs rkck(rk5)');
error_yrkck = np.absolute(yrkck-analytic)
error_rk = np.absolute(yrk-analytic)
newAx=plt.subplot(212)
l2=newAx.plot(timeVec,error_yrkck,'g-',label='rkck')
l3=newAx.plot(timeVec,error_rk,'m-',label='rk')
newAx.set_xlabel('time (seconds)')
newAx.legend(loc='best')
newAx.set_title("Error ");
```

### Problem Coding A

```python
import json
from numlabs.lab4.example.do_example import get_init
#
# specify the derivs function
#
def derivs(coeff, y):
  f=np.empty_like(y) #create a 2 element vector to hold the derivative
  f[0]=y[1]
  f[1]= -coeff.c1*y[1] - coeff.c2*y[0]
  return f

def heuns(coeff,y,derivs):
    k1 = coeff.dt * derivs(coeff,y)
    k2 = coeff.dt * derivs(coeff,y + (2/3 * k1))
    f = y + (1/4)*k1 + (3/4)*k2
    return f

# first make sure we have an input file in this directory


coeff=get_init()

# integrate and save the result in savedata

time=np.arange(coeff.t_beg,coeff.t_end,coeff.dt)
y=coeff.yinitial
nsteps=len(time) 
savedata=np.empty([nsteps],np.float64)
for i in range(nsteps):
    y=heuns(coeff,y,derivs)
    savedata[i]=y[0]

theFig,theAx=plt.subplots(1,1,figsize=(8,8))
theAx.plot(time,savedata,'o-')
theAx.set_title(coeff.plot_title)
theAx.set_xlabel('time (seconds)')
theAx.set_ylabel('y0');
```

An explicit two stage Runge-Kutta method can be written as:

\begin{align}
k_1 =&  h f(y_n,t_n)\\
k_2 =& h f(y_n+b_{21}k_1, t_n+a_2h) \\
y_{n+1} =& y_n + c_1k_1 +c_2k_2
\end{align}

Heun's method:

$$
  \begin{array}{l}
    k_1 =  h f(y_n,t_n)\\
    k_2 = h f(y_n+\frac{2}{3}k_1, t_n+\frac{2}{3}h)\\
    y_{n+1} = y_n + \frac{1}{4}k_1 + \frac{3}{4}k_2
  \end{array}
$$

Midpoint method:
$$
\begin{aligned}
&k_{1}^*=h f\left(y_{n}, t_{n}\right)\\
&k_{2}^*=h f\left(y_{n}+\frac{1}{2} k_{1}^*, t_{n}+\frac{1}{2} h\right)\\
&y_{n+1}=y_{n}+k_{2}^*
\end{aligned}
$$

We can show that $\frac{1}{4}k_1 + \frac{3}{4}k_2 = k_2^*$ , where $^* $ indicates the coefficient using the midpoint method.

Let $f(y,t) = t - y  + 1$ 

LHS: $$k_2^* = t+\frac{1}{2}h-\left(y+\frac{1}{2}\left(t-y+1\right)\right)+1 = \frac{1}{2}t+\frac{1}{2}h-\frac{1}{2}y+\frac{1}{2} $$

RHS: 

$$
\frac{1}{4} k_{1}^*+\frac{3}{4} k_{2}^* = \frac{1}{4}\left( t-y+1 \right) + \frac{3}{4}\left( t+\frac{2}{3}h-\left(y+\frac{2}{3}\left( t-y+1 \right)\right)+1\right) = \frac{1}{2}t+\frac{1}{2}h-\frac{1}{2}y+\frac{1}{2} = k_2^*
$$


For $f(y, t)=t-y+1$, the midpoint and Heun's method are the same.
