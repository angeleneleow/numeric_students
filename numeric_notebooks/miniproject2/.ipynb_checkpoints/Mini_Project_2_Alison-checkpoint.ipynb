**Starting with the daisy growth equations from Lab 5 and adding a rabbit population:**

White Daisies:
$$
\frac{dA_w}{dt} = A_w ( \beta_w (1- A_w - A_b) - \chi - D_w A_r)
$$

Black Daisies:
$$
\frac{dA_b}{dt} = A_b ( \beta_b (1- A_w - A_b) - \chi - D_b A_r)
$$

**Equations for Rabbit and Fox Population Growth Rate**

Rabbits:
$$
\frac{dA_r}{dt} = A_r ( \beta_r (A_w + A_b) - \chi - D_r A_f)
$$

Foxes: 
$$
\dfrac{dA_f}{dt} = A_f ( \beta_f A_r - \chi)
$$

Where:

$ A_i $ =  fractions of the total planetary area covered by population *i* <br>
$ \beta _i $ =  growth rate of population *i* <br>
$ \chi $ = natural death rate <br>
$ D_i $ = death rate (by predators) for population *i*
